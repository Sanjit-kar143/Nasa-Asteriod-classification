The data is about Asteroids - NeoWs.
NeoWs (Near Earth Object Web Service) is a RESTful web service for near earth Asteroid information. With NeoWs a user can: search for Asteroids based on their closest approach date to Earth, lookup a specific Asteroid with its NASA JPL small body id, as well as browse the overall data-set.

```python
nasa_data= nasa_data.drop(['Est Dia in KM(max)', 'Est Dia in M(min)', 'Est Dia in M(max)', 'Est Dia in Miles(min)'
             ,'Est Dia in Miles(max)', 'Est Dia in Feet(min)', 'Est Dia in Feet(max)', 
             'Relative Velocity km per hr', 'Miles per hour', 'Miss Dist.(lunar)', 
             'Miss Dist.(kilometers)', 'Miss Dist.(miles)'], axis = 1)
nasa_data.head()
```

```python
#drop the false columns
nasa_data.drop([False],axis=1,inplace=True)
nasa_data.head()
```

## Model Buliding

```python
x = nasa_data.drop([True],axis=1)
y = nasa_data[True].astype(int)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0 , test_size = 0.3)
```

## XGBoost Classifier 
  it is a decision-tree-based ensemble Machine Learning algorithm use in gradient boosting framework. In prediction problems involving unstructured data. A wide range of applications: Can be used to solve regression, classification, ranking, and user-defined prediction problems.

```python
from xgboost import XGBClassifier
from matplotlib import pyplot
from xgboost import plot_importance

xbg_model = XGBClassifier()
xbg_model.fit(x_train, y_train)
plot_importance(xbg_model)
pyplot.show()
```

```python
predictions = xbg_model.predict(x_test)
acc = accuracy_score(y_test,predictions)
acc*100
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
